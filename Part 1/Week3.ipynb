# Autonome Systeme

Die explizite Darstellung einer gewöhnliche Differentialgleichung sieht folgendermaßen aus:

$$
\dfrac{d}{dt}\vec{x} = \vec{f}(\vec{x},t)
$$

Ist die Funktion $\vec{f}(\vec{x},t)$ nicht direkt von der unabhängigen Varaiblen $t$ abhängig, sprechen wir von einer autonomen Differentialgleichung oder einem autonomen System.

$$
\dfrac{d}{dt}\vec{x} = \vec{f}(\vec{x})
$$

Gibt es eine unabhängige Variable $t$ und eine abhängige Variable $x$, so schreiben wir:

$$
\dfrac{d}{dt}x = f(x)
$$


Gibt es eine unabhängige Variable $t$ und zwei abhängige Variablen $x$  $y$, so schreiben wir:

$$
\dfrac{d}{dt}
\begin{pmatrix}
x \\ y
\end{pmatrix}
=
\begin{pmatrix}
f_1(x,y) \\
f_2(x,y)
\end{pmatrix}
$$

***Bermerkung***: Die Neuronen-Modelle Leaky-Integrator, Quad-Leaky-Integrate-And-Fire-Neuron und Huxley-Hodgkin sind autonome Systeme, solange es keine Zeitabhängigen Strominput gibt.

$$
J=
\begin{pmatrix}
\frac{\partial f_1}{\partial x} & 
\frac{\partial f_1}{\partial y}\\
\frac{\partial f_2}{\partial x} & 
\frac{\partial f_2}{\partial y}
\end{pmatrix}
$$
Die Art der Fixpunkte hängt von den ***Eigenwerten der Jakobi-Matrix an den Fixpunkten*** ab. Ist der Realteil der beiden Eigenwerte negativ ist der Fixpunkte stabil. 

    Das ist äquivalent zu
$$
\text{trace}(J) = \dfrac{\partial f_1}{\partial x_1} + \dfrac{\partial f_2}{\partial x_2} < 0, 
$$
$$
\text{det}(J) = \dfrac{\partial f_1}{\partial x_1}\dfrac{\partial f_2}{\partial x_2}-\dfrac{\partial f_1}{\partial x_2}\dfrac{\partial f_2}{\partial x_1}> 0.
$$

Siehe: https://en.wikipedia.org/wiki/Stability_theory

## Bemerkung:  Gewöhnliche DGLn höherer Ordnung
Ist die Ordnung einer gewöhnlichen DGL höher als eins. Lässt sie sich in ein System aus Gleichungen erster Ordnung umformen. Beispiel 2te Ordnung, 1 abhängige Variable:
$$
\dfrac{d^2x}{dt^2}=f(x,t)
$$
Wir definierten eine weitere Variable $y$ und erhalten eine zweite Gleichung
$$
y := \dfrac{dx}{dt}
$$
Durch Einsetzen erhalten wir ein System aus zwei Gleichungen.
$$
(1)~~~\dfrac{dy}{dt}=f(x,t) 
$$
$$
(2)~~~\dfrac{dx}{dt} = y 
$$

Wir definieren nun
$$
f(x,t) = f_2(x,y,t)
$$
$$
y = f_1(x,y,t) 
$$

und erhalten

$$
(1)~~~\dfrac{d}{dt}x=f_1(x,y,t)
$$
$$
(2)~~~\dfrac{d}{dt}y = f_2(x,y,t) 
$$

# Lösung 1

Dies ist keiner ausführliche Lösung der Aufgabe, sondern dient nur zur Kontrolle der eigenen Ergebnisse!

1. DGL autonom und erster Ordnung
2. x abhängigen und t unabhängig
3. Nicht-linear, Homogenität existiert nur im Kontext linearer DGLn.

4. dynamische Form:

$$
 \dfrac{dV}{dt} = -\dfrac{(V-E)}{\tau}+\dfrac{\Delta}{\tau} \cdot \exp(\dfrac{V-V_T}{\Delta}) + \dfrac{I}{g\tau}
$$


5. Visualsierung siehe unten.
6. Bestimmung numerisch siehe unten
7. Erster Fixpunkt $V_1$ stabil, zweiter $V_2$ instabil ($V_1<V_2$)
8. Bifurkationsparameter $I$
9. Numerische Lösung der DGL für verschiedene Randbedingungen und Werte des Bifurkationsparameter siehe unten


```python
import numpy as np
import numpy.linalg as linalg
import matplotlib.pyplot as plt
from scipy.integrate import odeint  
from scipy.optimize import root
from scipy.optimize import minimize


def solve(func, x0, t, x_reset=-0.08, x_max=0.01):
    """
    Solves the initial value problem
    dx/dt = func(x, t)  
    for LIF-Neurons where the voltage is reset to model a spike
    
    Parameters
    ----------
    func : callable(x, t) 
    x0 : float
        Initial condition on x 
    t : array
        A sequence of time points for which to solve for x. The initial
        value point should be the first element of this sequence.
        This sequence must be monotonically increasing or monotonically
        decreasing; repeated values are allowed.
    x_reset, x_max: float
        if x grows larger than x_max, then x is reset to x_reset 
    
    """
    x = np.zeros(np.shape(t))
    x[0] = x0
    for i, t_i in enumerate(t[:-1]):
        j = i+1
        t_j=t[j]
        x_j = odeint(f, x[i],[t_i,t_j])[1]
        if x_j>x_max:
            x_j=x_reset
        x[j] = x_j
    return x
```

### Bemerkung: Numerisch Extremstellen (Minima) von Funktionen finden.

```python
# from scipy.optimize import minimize

def f_square(x):
    return (x+0.123)**2-1.45

xguess = 0.  # initial location for search
minimum = minimize(f_square, x0=xguess).x[0]
print('Location of minimum: ', minimum)
print('Value of function at minimum: ', f_square(minimum))
```

```python
#model parameters
tau = 0.02
E = -0.07
Delta = .01
V_T = -0.04
I = .0 # <- Injected current, bifurcation parameter 
g = 0.3

def f(V,t=None):
    return -(V-E)/tau + Delta/tau*np.exp((V-V_T)/Delta) + I/g/tau

# find bifurcation point
############################
vmin = minimize(f,0).x[0]
i_bif = -f(vmin)*g*tau
print('Bifurcation current: ', i_bif)

# set injected current
#############################
I = 0.#i_bif * 1.2 # <- Injected current, bifurcation parameter 

# find fixpoints as roots of f
############################
fixpoint1 = root(f, x0=0.).x[0]
fixpoint2 = root(f, x0=-0.1).x[0]
print('Fixpunkte:', fixpoint1, fixpoint2)

# solve ode
############################
t = np.arange(0. ,0.1,1.e-6)
v0 = vmin
x_num = solve(f, v0, t)

# display results
#############################
fig= plt.figure(dpi=150, figsize=(5.,2.))
ax1 = fig.add_subplot(121)
ax1.plot(t,x_num, label='numerical solution', ls='--')
#ax1.plot([t[0]],[x_num[0]], 'ro', label='initial value')
ax1.legend(fontsize=8, frameon=False)
ax1.set_xlabel('t')
ax1.set_ylabel('x(t)')
ax1.set_ylim(-0.1, 0.0)
ax1.set_xlim(-0.01,0.11)

ax2 = fig.add_subplot(122)
x = np.linspace(-0.1,0.01,100)
f_x = f(x,np.zeros(100))
ax2.plot([fixpoint1, fixpoint2],[0,0],'kx')

ax2.plot(x,f_x)
ax2.plot(x,np.zeros(100),'k--', lw=1)
ax2.set_ylabel('f(x)')
ax2.set_xlabel('x')
ax2.set_ylim(-5,5)
plt.tight_layout()
fig.show()
```

5. Visualsieren der Funktion $(f_1(x,y),f_2(x,y)^T$ als Vektorfeld (Phasenraum):

```python
# b, eps > 0, a<0
a, b, eps = -2., .3, 1.

def f_1(x,y):
    return a*x-y

def f_2(x,y):
    return eps*(b*x-y)

y_min, y_max = -5, 5
x_min, x_max = -5, 5
res_x, res_y = 10, 10
x = np.linspace(x_min, x_max, 1000)
xx,yy = np.meshgrid(np.linspace(x_min,x_max,res_x),
          np.linspace(y_min,y_max,res_y))
dxdt, dydt = f_1(xx,yy), f_2(xx,yy)


fig = plt.figure(figsize=(5,5), dpi=150)
ax = fig.add_axes([0.15,0.15,.8,.8])
ax.quiver(xx,yy,dxdt,dydt, label=r'$(f_1(x,y),f_2(x,y))^T$')
ax.set_ylim(y_min, y_max)
ax.set_xlim(x_min, x_max)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
        
```

6. Bestimmen der Fixpunkte: Berechne die beiden Nullklinen und finde Fixpunkte als Schnittpunkte der Nullklinen. Visualsieren der Nullklinen im Phasenraum (siehe 5.)

```python
def nullcline_f1(x):
    return a*x
    
def nullcline_f2(x):
    return b*x
    
fig = plt.figure(figsize=(5,5), dpi=150)
ax = fig.add_axes([0.15,0.15,.8,.8])
ax.quiver(xx,yy,dxdt,dydt, label=r'$(f_1(x,y),f_2(x,y))^T$')
ax.plot(x,nullcline_f1(x), label=r'Nullkline 1 $\dfrac{dx}{dt}$')
ax.plot(x,nullcline_f2(x), label=r'Nullkline 2 $\dfrac{dy}{dt}$')
ax.plot([0],[0], 'ro', label='Fixpunkt')
ax.set_ylim(y_min, y_max)
ax.set_xlim(x_min, x_max)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
```

$$
J=
\begin{pmatrix}
a & 
-1\\
\epsilon b & 
-\epsilon
\end{pmatrix}
$$

A. Stabilitätsanalyse mit Spur und Derminante:

$$
\text{trace}(J) = a - \epsilon < 0, 
$$

$$
\text{det}(J) = -a\epsilon+\epsilon b = \epsilon(b-a)> 0.
$$

B. Stabilitätsanalyse über Eigenwerte von J für den Fall $a=-1$ und $\epsilon=b=1$
    
    Das charakteristische Polynom ist in diesem fall $\lambda^2+2\lambda+2=0$
    
    Die Eigenwerte sind also $\lambda_1=1-i$ und $\lambda_2=1+i$



Der Fixpunkt also stabil

8. Bifurkationsanalyse: Für den Fall $b<a$ wird die Determinante der Jakobimatrix negative und der Fixpunkt deshalb instabil.

9. Numerical solution and visualization

```python
a,b,eps = -1., 1., 1.

def f(x,t=None):
    x1,x2 = x
    return f_1(x1,x2), f_2(x1,x2)

xy0 = -4., -4.
t = np.linspace(0,5,100)
xy = odeint(f, xy0, t).transpose()

fig = plt.figure(figsize=(5,5), dpi=150)
ax = fig.add_axes([0.15,0.15,.8,.8])
ax.quiver(xx,yy,dxdt,dydt, label=r'$(f_1(x,y),f_2(x,y))^T$')
ax.plot(x,nullcline_f1(x), label='Nullkline 1')
ax.plot(x,nullcline_f2(x), label='Nullkline 2')
ax.plot([0],[0], 'ro', label='Fixpunkt')
ax.plot(*xy, label='Solution')
ax.set_ylim(y_min, y_max)
ax.set_xlim(x_min, x_max)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
```
