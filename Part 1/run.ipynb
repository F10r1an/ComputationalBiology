# Simulate dynamical system with two dependent variables
Example for FitzHugh-Nagumo model

```python
from AutoSys2D import FitzHugh_Nagumo


FHN_std_parameters = {'a': 0.7, 
                      'b': 0.8,
                      'tau': 12.5,
                      'i': 1.,
}

FHN_cell = FitzHugh_Nagumo(x0=-1.5, y0=-1., parameters=FHN_std_parameters, run_time=200, front_end=True)
FHN_cell.run()
```

# Show animation of solution

```python
%matplotlib notebook
FHN_cell.notebook_animation(equal_range=True)
```

# Create video

```python
# crate video
# this may take a while...
video = FHN_cell.webapp_animation(equal_range=True)
vid = video.to_html5_video()
```

```python
# display video in notebook
from IPython import display 
# embedding for the video 
html = display.HTML(vid) 
  # draw the animation 
display.display(html) 
```

```python
# save video to disk
# may also take a while...
import matplotlib.animation as animation 
writervideo = animation.FFMpegWriter(fps=60) 
video.save('FHN.mp4', writer=writervideo) 
```
